#### 4. Variables seem to have incorrect type and inconsistent formatting. You need to fix them: 

   **1. Size column has sizes in Kb as well as Mb. To analyze, you’ll need to convert these to numeric.**
   - Extract the numeric value from the column
   - Multiply the value by 1,000, if size is mentioned in Mb**

```python
play_store_data['Size']
```

```python
def size_change(size):
    size_num = size[:-1]
    if 'M' in size:
        return float(size_num) * 1000000
    elif 'k' in size:
        return float(size_num) * 1000
    else:
        return None
    
play_store_data['Size']=play_store_data['Size'].apply(size_change)
```

#### 8.Data Preprocessing

```python
play_store_data['Installs']=play_store_data['Installs'].apply(func=np.log1p)
play_store_data['Reviews']=play_store_data['Reviews'].apply(func=np.log1p)

play_store_data.hist(column=['Installs','Reviews'])
```

```python
import copy
# copying the original dataset to a temp dataset
new_data = copy.deepcopy(play_store_data)
new_data
```

**Drop columns App, Last Updated, Current Ver, and Android Ver. These variables are not useful for our task.**

```python
new_data.drop(['App', 'Last Updated', 'Current Ver', 'Android Ver'], axis =1,inplace=True)
```

```python
new_data.info()
```

```python
# creating dummies for Content Rating and dropping the column from the table
content_rating_variable=pd.get_dummies(new_data['Content Rating'],drop_first=True)
new_data1 = pd.concat([new_data,content_rating_variable], axis=1)
new_data1.drop('Content Rating', axis=1,inplace=True)
new_data1.head()
```

```python
# creating dummies for Category and dropping the column from the table
category_variable=pd.get_dummies(new_data1['Category'],drop_first=True)
new_data1 = pd.concat([new_data1,category_variable], axis=1)
new_data1.drop('Category', axis=1,inplace=True)
new_data1.head()
```

```python
# creating dummies for Genres and dropping the column from the table
category_variable=pd.get_dummies(new_data1['Genres'],drop_first=True)
new_data1 = pd.concat([new_data1,category_variable], axis=1)
new_data1.drop('Genres', axis=1,inplace=True)
new_data1.head()
```

```python
#dropping the type column
new_data1.drop('Type',axis=1,inplace=True)
new_data1.head()
```

```python
#new_data1.drop('Reviews',axis=1,inplace=True)
```

```python
new_data1.info()
```

Splitting Dataset into training and testing dataset

```python
X=new_data1.iloc[:,1:]
y=new_data1.iloc[:,:1]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.30, random_state=1)
```

Apply linear regression to the training set

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
model=regressor.fit(X_train, y_train)
```

```python
y_pred=regressor.predict(X_test)
```

Find the value of adjusted R2 , R2 score and RMSE value

```python
from statsmodels.api import OLS
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error as ms
summ=OLS(y_train,X_train).fit()
summ.summary()
```

```python
print('R2_Score=',r2_score(y_test,y_pred))
print('Root_Mean_Squared_Error(RMSE)=',np.sqrt(ms(y_test,y_pred)))
```

**THE VALUE OF adjusted R2 IS AROUND 0.985 WHICH IS A VERY GODD VALUE, MORE THIS VALUE NEARER TO 1 BETTER IS THE CORRELATION BETWEEN PREDICTED AND TEST VALUES**
